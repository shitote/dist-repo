### **Steps in modeling**

1. ***Creating a model*** - Define the input and the output layers, as well as the hidden layers of a deep learning model
2. ***Compiling a model*** = Defin the loss function (in that the function that will the our model who wrong it is performing) and the optimizer (the optimizer tells the model how to improve the learing patterns) The the evaluation metrixs(which we can use to predict how our model is performing)
3. ***Fitting a model*** - Letting the model to try and find patterns between x and y

```python
# Set random seed
x = np.reshape(x, (x.shape[0], 1))
tf.random.set_seed(42)

# Create a model using the syquential api.
model = tf.keras.Sequential([
    tf.keras.layers.Dense(1),

])

# Compile the model
model.compile(loss=tf.keras.losses.mae,
              optimizer= tf.keras.optimizers.SGD(), #This is the maighty stochastic gradient discent
              metrics =['mae'])


# Fit the model
model.fit(x, y, epochs=5);
```

```python
# Make a function for poting.
def Plot_predictions(train_data=x_train,
                     train_labels=y_train,
                     test_data=x_test,
                     test_labels=y_test,
                     predictions=y_pred):
  """
  plot training data, testing data and compare it to the predictions.
  """
  plt.figure(figsize=(10, 7))
  # Plot training data in blue
  plt.scatter(train_data, train_labels, c='b', label="Training data")
  # plot the testing data on the same plot
  plt.scatter(test_data, test_labels, c="orange", label="testing data")
  # Plot the predictions in purple
  plt.scatter(test_data, predictions, c='purple', label='predictions')
  # Show the lagent
  plt.legend()
```

```python
# Function to do the calculations of the mae ans mse
def mae(y_true, y_pred):
  return tf.metrics.mean_absolute_error(y_true=y_true,
                                        y_pred=tf.squeeze(y_pred))

def mse(y_true, y_pred):
  return tf.metrics.mean_squared_error(y_true=y_true,
                                       y_pred=tf.squeeze(y_pred))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

# Read the insurance data frame frome the git repository.
insurance = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
insurance
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split

# Create a column transformer
ct = make_column_transformer(
    (MinMaxScaler(), ['age', 'bmi', 'children']),
    (OneHotEncoder(handle_unknown='ignore'), ['sex', 'smoker', 'region'])
)

# Create x and y
x = insurance.drop('charges', axis=1)
y = insurance.charges

# Split the data into training and testing data.
x_train, x_test, y_train, y_test = train_test_split(x, y,
                                                    test_size=0.2,
                                                    random_state=42)

# Fit the column traonsformer to our training data
ct.fit(x_train)

# Transform training and test data with normalization (MinmaxScaler) and onehotencoder.
x_train_normal = ct.transform(x_train)
x_test_normal = ct.transform(x_test)
```

```python
# Now see how our data looks like after transormation.
x_train.loc[0]
```

```python
x_train_normal[0]
```

Beautiful! Now that the data hase been normalized and onehotEcoded it is time to build a neural net to fing patterns in the data

```python
# Set the rundom seed
tf.random.set_seed(42)

# Create the model.
insurance_model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

# Compile the model.
insurance_model_4.compile(loss='mae',
                          optimizer='Adam',
                          metrics=['mae'])

# Fit the model with the normalized traing data.
insurance_model_4.fit(x_train_normal, y_train, epochs=100);
```

```python
# Evaluate our insurance model trained on normalized data
insurance_model_4.evaluate(x_test_normal, y_test)
```
